```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python


# Step a: Create a figure object called fig
fig = plt.figure()

# Step b: Use add_axes to add an axis to the figure canvas at [0,0,1,1]
# This specifies that the axis should take up the entire figure
ax = fig.add_axes([0, 0, 1, 1])

# Sample data for plotting
x = [0,20,40,60,80,100]
y = [2*i for i in x]
ax.set_yticks([0,50,100,150,200])

# Step c: Plot (x, y) on that axes and set the labels and titles
ax.plot(x, y)
ax.set_xlabel('X-axis Label')  # Set the x-axis label
ax.set_ylabel('Y-axis Label')  # Set the y-axis label
ax.set_title('Plot Title')      # Set the plot title

# Show the plot
plt.show()
```

```python
df=pd.read_csv('./company_sales_data.csv')
df.head()
```

```python
month=df['month_number']
profit=df['total_profit']
fig=plt.figure()
ax=fig.add_axes([0, 0, 1, 1])

# Plot the data
ax.plot(month, profit)

# Set labels and title
ax.set_xlabel('Month Number')  # X-axis label
ax.set_ylabel('Total profit')  # Y-axis label
ax.set_title('Total Profit Over Time')  # Plot title
# Show the plot
plt.show()
```

Q4

```python
month=df['month_number']
profit=df['total_units']
fig=plt.figure()
ax=fig.add_axes([0, 0, 1, 1])

# Plot the data
ax.plot(month, profit,linestyle='--', color='red', marker='o',markerfacecolor='red' , markersize=8,linewidth=3,label='Sold Units')
ax.legend(loc='lower right')
# Set labels and title
ax.set_xlabel('Month Number')  # X-axis label
ax.set_ylabel('Sold Units Number')  # Y-axis label
ax.set_title('Total Profit Over Time')  # Plot title
# Show the plot
plt.show()
```

Additonal 1

```python
adv = df
fig,ax=plt.subplots()
products=adv.columns[1:7]
for product in products:
    ax.plot(adv['month_number'],adv[product],marker='o',label=product)
ax.set_xlabel('Month Number')  # X-axis label
ax.set_ylabel('Sold Units Number')  # Y-axis label
ax.set_title('Total Profit Over Time')  # Plot title
plt.legend(title='Products',loc='upper left')
plt.xticks(rotation=45)
plt.show()
```
