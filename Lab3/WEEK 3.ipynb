# Q2(CALCULUS)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset from the CSV file
df = pd.read_csv('ml_lab_course_data.csv')

# Features and target variable
X = df['Study Time (hours)'].values
y = df['Score (out of 10)'].values

# Gradient Descent parameters
alpha = 0.01  # Learning rate
iterations = 1000  # Number of iterations

# Initialize parameters
B0 = 0
B1 = 0
n = len(X)

# Gradient Descent
for _ in range(iterations):
    # Compute predictions
    y_pred = B0 + B1 * X
    
    # Compute gradients
    dB0 = - (1/n) * np.sum(y - y_pred)
    dB1 = - (1/n) * np.sum((y - y_pred) * X)
    
    # Update parameters
    B0 -= alpha * dB0
    B1 -= alpha * dB1

# Predict responses
y_pred_final = B0 + B1 * X

# Calculate RMSE
rmse = np.sqrt(np.mean((y - y_pred_final) ** 2))

# Print results
print(f"Using gradient descent with partial derivatives:")
print(f"Intercept (B0): {B0}")
print(f"Slope (B1): {B1}")
print(f"Root Mean Square Error (RMSE): {rmse}")

# Create scatter plot of the data points
plt.scatter(X, y, color='red', label='Data Points')

# Create a line for the predicted values
x_range = np.linspace(X.min(), X.max(), 100)
y_range = B0 + B1 * x_range

# Plot the regression line
plt.plot(x_range, y_range, color='blue', label='Regression Line')

# Labeling the plot
plt.xlabel('Study Time (hours)')
plt.ylabel('Score (out of 10)')
plt.title('Study Time vs. Score with Regression Line (Gradient Descent)')
plt.legend()

# Show the plot
plt.show()
```
